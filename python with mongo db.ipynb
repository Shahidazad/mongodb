```python
# condition 
# select * from table where condition
for i in empinfo.find({'ln':'kil'}):
    print(i)
```

```python
for i in empinfo.find({'ln':{'$in':['kiql','kiqwl']}}):# filter
    print(i)
```

```python
for i in empinfo.find({'fn':'sha','age':{'$lt':122}}):# 2 condition(and operator)
    print(i)
```

```python
for i in empinfo.find({'$or':[{'ln':'kil'},{'age':1222}]}):# or condition
    print(i)
```

```python
for i in empinfo.find({'$and':[{'ln':'kil'},{'age':1222}]}):# and condition
    print(i)
```
